```python
from konlpy.tag import Okt
from collections import Counter
import csv
```

```python
okt = Okt()
```

```python
with open("F:/eunjin/NLP/data/newslist.csv", "r", encoding='utf-8') as f:
    csvr = csv.reader(f)
    for news in csvr: # x1000
        noun = okt.nouns(str(news))
        count = Counter(noun)

        for i, j in enumerate(noun):
            if len(j) < 2:
                noun.pop(i)
            elif (j=="기자" or j=="연합뉴스"):
                noun.pop(i)
    
        noun_list = count.most_common(10) # noun_list.append(count.most_common(10))
        for i in noun_list:
            print(i)
            
```
